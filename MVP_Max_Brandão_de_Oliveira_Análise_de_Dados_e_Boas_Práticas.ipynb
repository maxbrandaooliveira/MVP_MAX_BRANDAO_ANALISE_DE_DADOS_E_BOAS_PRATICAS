A segurança no trânsito é um aspecto de extrema relevância mundial. De acordo com a Organização Mundial da Saúde (OMS), estima-se que, a cada ano, cerca de 1,25 milhões de pessoas morrem no trânsito e até 50 milhões sobrevivem com ferimentos e sequelas, resultando em custos pessoais, sociais e econômicos. Por exemplo, meu pai faleceu em um acidente de trânsito em 1989 quando o mesmo tinha 26 anos de idade e eu tinha apenas 1 ano.

Os acidentes de trânsito são uma das principais causas de morte entre jovens com idades entre 15 e 29 anos (OMS, 2018). Em 2020, para discutir estratégias para reduzir as mortes, a OMS organizou a terceira Conferência Ministerial Global sobre Segurança no Trânsito (OMS, 2020). No Brasil, de acordo com os dados divulgados pela Confederação Nacional do Transporte (CNT), o número de acidentes nas rodovias federais brasileiras aumentou em 1,6% em 2021 em relação a 2020, totalizando 64.452 casos. Além disso, houve um aumento de 2% no número de mortes, passando de 5.287 em 2020 para 5.391 em 2021, de acordo com os dados publicados pela Polícia Rodoviária Federal (PRF) (TRANSPORTE, CNT, 2021).

## Atributos do Dataset

Cada dataset possui **TRINTA** atributos originalmente declarados na seguinte forma:

- ***ano*** (int64): Identifica o ano do sinistro;
- ***id***	(float64):	Um identificador único para cada registro no conjunto de dados, permitindo fácil referência e rastreamento de entradas individuais;
- ***data_inversa***	(str):	A data em que o evento ocorreu, fornecendo um contexto temporal para os dados;
- ***dia_semana***	(str):	O horário exato em que o evento ocorreu, permitindo uma análise temporal detalhada;
- ***horario***	(str):	O horário exato em que o evento ocorreu, permitindo uma análise temporal detalhada;
- ***uf***	(str):	Representa o estado onde o evento ocorreu, o que é essencial para a análise regional;
- ***br***	(int64):	O número da rodovia federal associado ao evento, útil para identificar rotas específicas;
- ***km***	(str):	O marcador quilométrico na rodovia onde o incidente ocorreu fornece informações precisas sobre a localização;
- ***municipio***	(str):	O município onde o evento ocorreu, auxiliando na análise e na elaboração de relatórios locais;
- ***causa_acidente***	(str):	Descreve a causa do acidente, o que é fundamental para compreender os fatores que contribuíram para ele;
- ***tipo_acidente***	(str):	Classifica o tipo de acidente, ajudando a identificar tendências e cenários comuns;
- ***classificacao_acidente***	(str):	Classifica a gravidade do acidente, o que é importante para avaliar o impacto e as necessidades de resposta;
- ***fase_dia***	(str):	Indica a fase do dia (por exemplo, manhã, tarde, noite) em que o evento ocorreu, útil para análises baseadas no tempo;
- ***sentido_via***	(str):	Descreve a direção do fluxo de tráfego na estrada no momento do evento, o que pode influenciar a dinâmica do acidente;
- ***condicao_metereologica***	(str):	Descreve detalhadamente as condições meteorológicas no momento do evento, que podem ser um fator significativo na ocorrência de acidentes;
- ***tipo_pista***	(str):	Especifica o tipo de superfície da estrada, por exemplo, asfalto, cascalho, onde o evento ocorreu, relevante para avaliações de segurança;
- ***tracado_via***	(str):	Descreve o traçado da estrada por exemplo, reta, curva, o que pode influenciar as taxas de acidentes;
- ***uso_solo***	(str):	Indica o tipo de uso do solo na área ao redor do evento, o que pode fornecer contexto para os padrões de tráfego;
- ***pessoas***	(int):	Número total de pessoas envolvidas no evento é importante para entender a dimensão do incidente;
- ***mortos***	(int64):	Número de fatalidades resultantes do evento é uma métrica crucial para avaliar a gravidade dos acidentes;
- ***feridos_leves***	(int64):	Contabiliza o número de indivíduos com ferimentos leves, o que é útil para compreender o impacto nas vítimas;
- ***feridos_graves*** (int64): Número de feridos graves, que necessitam de atendimento de urgência e oferem risco aos pacientes;
- ***ilesos*** (int64): Contabiliza o número de envolvidos que não sofreram lesões no sinistro;
- ***ignorados*** (int64): Indica o número de casos ignorados no sinistro;
- ***feridos*** (int64): Contabiliza o número de feridos envolvidos no sinistro;
- ***veiculos*** (int64): Indica a quantidade de veículos envolvidos no sinistro;
- ***latitude*** (float64): Indica no mapa a latitude do sinistro;
- ***longitude*** (float64): Indica no mapa a logitude do sinistro;
- ***regional*** (str):  Indica a codificação de qual estado pertence a delegacia que atendeu ao sinistro;
- ***delegacia*** (str): Indica qual delegacia foi responsável por atender e registrar a ocorrência do sinistro, já que em cada estado existe diversas delegacias;
- ***uop*** (str): Se refere a qual unidade de uma delegacia foi responsável pelo atendimento do sinistro.

De posse dessas variáveis originais, teremos foco na presença de óbito no sinistro, bem como na condição de visibilidade no momento da ocorrência. Cada regristo nos permitirá extrair outras informações, como mês, turno, visibilidade e outras.

```python
# Obtendo os arquivos .csv zipados disponibilizados no link
files = { 2015: "1DyqR5FFcwGsamSag-fGm13feQt0Y-3Da",
          2016: "16qooQl_ySoW61CrtsBbreBVNPYlEkoYm",
          2017: "1HPLWt5f_l4RIX3tKjI4tUXyZOev52W0N",
          2018: "1cM4IgGMIiR-u4gBIH5IEe3DcvBvUzedi",
          2019: "1pN3fn2wY34GH6cY-gKfbxRJJBFE0lb_l",
          2020: "1esu6IiH5TVTxFoedv6DBGDd01Gvi8785",
          2021: "12xH8LX9aN2gObR766YN3cMcuycwyCJDz",
          2022: "1PRQjuV5gOn_nn6UNvaJyVURDIfbSAK4-",
          2023: "1-WO3SfNrwwZ5_l7fRTiwBKRw7mi1-HUq",
          2024: "14lB0vqMFkaZj8HZ44b0njYgxs9nAN8KO",
          2025: "1-G3MdmHBt6CprDwcW99xxC4BZ2DU5ryR"}

# Arquivo temporário para agrupar todos em um único arquivo
tmp_dir = tempfile.mkdtemp()
dfs = []
encodings_to_try = ["utf-8", "latin1", "cp1252"]

for year, fid in files.items():
    dest = os.path.join(tmp_dir, f"datatran{year}.zip")
    url = f"https://drive.google.com/uc?id={fid}"
    print(f"Baixando {year}...")
    gdown.download(url, dest, quiet=True)
    if not os.path.isfile(dest):
        print(f"Aviso: download falhou para {year}, pulando.")
        continue

    try:
        with zipfile.ZipFile(dest, "r") as z:
            names = z.namelist()
            csv_candidates = [n for n in names if n.lower().endswith(".csv") and str(year) in n]
            if not csv_candidates:
                csv_candidates = [n for n in names if n.lower().endswith(".csv")]
            if not csv_candidates:
                print(f"Aviso: nenhum CSV em {dest}, pulando.")
                continue
            csv_name = csv_candidates[0]

            # tenta com vários encodings
            df = None
            for enc in encodings_to_try:
                try:
                    with z.open(csv_name) as f:
                        df = pd.read_csv(f, sep=';',decimal=',', encoding=enc, low_memory=False)
                    print(f"Lido {year} com encoding {enc}, linhas: {len(df)}")
                    break
                except Exception as e:
                    print(f"Falha leitura {year} com {enc}: {e}")
                    df = None

            if df is None:
                print(f"Não foi possível ler {csv_name} em {year}, pulando.")
                continue

            df["ano"] = year
            dfs.append(df)

    except zipfile.BadZipFile:
        print(f"Arquivo zip inválido: {dest}")
    except Exception as e:
        print(f"Erro no ano {year}: {e}")

if not dfs:
    raise SystemExit("Nenhum CSV lido. Verifique os arquivos.")

df_all = pd.concat(dfs, ignore_index=True)
cols = ["ano"] + [c for c in df_all.columns if c != "ano"]
df_all = df_all[cols]
out = "datatran_2015_2025.csv"
df_all.to_csv(out, index=False, encoding="utf-8", sep=';')

```

# Hipótese 2: O número de óbitos tem tendência positiva por ano

Observando as quantidades de óbitos por ano, o comportramento não é previsível como apenas uma tendÊncia positiva. Somando os valores por ano, observamos de 2015 a 2018 um queda, que foi seguida por um aumento até 2024. Ao ajustar um modelo de regressão simples, o coeficiente angular foi negativo, evidenciando que o comportamento de tendência negativa na maior parte do período.

Devido o comportamento dos dados por ano, ajustamos um modelo quadrático com concavidade para cima (coeficiente positivo associado ao quadrado da variável). Percaba que o modelo foi bem mais ajustado do que o linear, como esperado. Veja que o polinômio descreveu bem o comportamento dos dados. ALém disso, todos os valores-p de cada coeficiente no modelo quadrático foram significativos, enquanto no modelo linear, ambos não foram significativos.

Esse comportamento, contradiz a hipótese levantada inicialmente que apontava uma tendência estritamente positiva. Apesar de que um modelo provavelmente mais adequado para os dados seria uma ARIMA (autoregressivo integrado de médias móveis) ou SARIMA  (ARIMA com componente sazonal, mas os dados de forma mensal não fornecem um tamanho de amostra suficiente para um ajuste de séries temporais. # Hipótese 4: Visibilidade x Óbito

Assim como na hipótese anterior, pelo teste qui-quadrado de Pearson, ao nível de 5% de significância, não rejeitamos a que há independência entre as variáveis, indicando que há relação entre o óbito e a visibilidade. Percebe-se que a visibilidade parcial apresentou a maior incidência de óbitos. Isto é, a condição boa permite o condutor evitar acidentes e tomar decisões corretas, enquanto a visibilidade ruim, faz com os condutores ruduzam a velocidade ou parem até ter uma condição razoável para voltar a dirigir. Já a parcial faz com que a confiança permaneça elevada, mas com baixas condições de visibilidade.

### Calculando o V de Cramér

A medida V de Cramér avalia a força de associação entre duas variáveis nominais. Um valor nulo indica nenhuma associação e um valor unitário indica uma associação perfeita. É calculado usando os resultados do teste qui-quadrado. Destaca-se que, apesar do teste qui-quadrado ter sido bastante significativo ao nível de 5%, o V de Cramér não apontou nehuma associação forte de acordo com a classificação abaixo:

- 0: Nenhuma associação;
- 0.1-0.3: Associação fraca;
- 0.3-0.5: Associação moderada;
- \> 0.5: Associação forte.

Das variáveis calculadas, as com maiores valores do V de cramér foram tipo de pista, fase do dia e unidade federativa, indicando que esssas variáveis têm uma relação mais forte do que comparada as demais. Isso pode ser visto e confirmado nos gráficos elaborados e na análise de correspondência multivariada.

```python
vars_cramer = ['mes','ano','uf','fase_dia_correto','dia_semana_correto','tipo_pista_correto','visibilidade']

results = []
for v in vars_cramer:
    ct = pd.crosstab(data_15_25[v], data_15_25['obito'])
    if ct.size == 0 or ct.shape[0] < 2 or ct.shape[1] < 2:
        results.append((v, np.nan, np.nan))
        continue
    chi2, p, dof, exp = chi2_contingency(ct)
    n = ct.values.sum()
    k = min(ct.shape)
    cramer_v = np.sqrt(chi2 / (n * (k - 1))) if n > 0 and (k - 1) > 0 else np.nan
    results.append((v, p, cramer_v))

pd.DataFrame(results, columns=['variavel','p_value','cramers_v']).set_index('variavel')
```

# Teste qui-quadrado para outras variáveis

Além das 4 hipóteses testadas, vamos observar de forma investigativa a relação com outras variáveis. Basicamente todas as variáveis selecionadas, tiveram relação entre si 2 a 2. Em outras palavras, a ocorrência de óbito no sinistro está relacionada à todas as variáveis selecionadas na análise final. Isto é, a ocorrência de óbtito no sinistro está relacionada ao ano, ao mês, à unidade federativa, ao dia da semana, à fase do dia e à visibilidade.

Não focando apenas no óbito, observamos que as demais variáveis se relacionam entre si. Por exemplo, UF e dia da semana, como fase do dia e visbilidade. Todas as combinações possíveis 2 a 2 mostraram-se dependentes. O gráfico mostra isso com muita clareza com a escala de cor, aonde é nítido que todos os valores-p tendem a zero na escala de cor (amarelo).

```python
## Vetor com variáveis categóricas
vars = ['ano','mes','uf','dia_semana_correto','fase_dia_correto','visibilidade','obito']
pmat = pd.DataFrame(np.nan, index=vars, columns=vars)

for i, a in enumerate(vars):
    for j, b in enumerate(vars):
        if i>=j:
            continue
        sub = data_15_25[[a,b]].dropna()
        if sub.empty:
            p = np.nan
        else:
            tbl = pd.crosstab(sub[a], sub[b])
            if tbl.size==0 or tbl.shape[0]<2 or tbl.shape[1]<2:
                p = np.nan
            else:
                _, p, _, _ = chi2_contingency(tbl, correction=False)
        pmat.at[a,b] = p
        pmat.at[b,a] = p

# Correção de Bonferroni
n_tests = np.triu(~pmat.isna(), 1).sum().sum()
alpha = 0.05
alpha_bonf = alpha / max(1, n_tests)
# Impressão dos valores-p np heat map
annot = pmat.copy().astype(object)
for i in annot.index:
    for j in annot.columns:
        v = annot.at[i,j]
        if pd.isna(v):
            annot.at[i,j] = '-'
        else:
            annot.at[i,j] = f"{v:.3e}" if v<0.001 else f"{v:.3f}"

mask = np.eye(len(vars), dtype=bool)

plt.figure(figsize=(10,10))
sns.heatmap(pmat.astype(float), annot=annot, fmt='', cmap='viridis_r', cbar_kws={'label':'p-valor'}, vmin=0, vmax=1, mask=mask)
plt.title(f'Valores-p do teste qui-quadrado (Bonferroni α={alpha_bonf:.3g})')
```

Com mais mais detalhes, podemos ver as tabelas de contingência de tais variáveis para observar melhor a incidência dentro de cada categoria dos atributos. Por ano, 2015 foi o ano com menor quantidade de disnistros com óbito. Por mês, julho foi o de maior proporção, que coincide com o período de férias escolares, o que leva ao aumento de viagens nesse período.

Quanto ao estado, os com maioes quantidades de óbito são Maranhão, Pará, Amazonas, Tocantins, Roraima, Alagoas e Bahia, do maior para o menor, respectivamente. Por outro lado, os com menores proponcões de sinistros com óbtio são Sãnta Catarina, Distrito Federal, São Paulo, Espírito Santo e Rio Grande do Sul. Percebe-se que as rodovias mais perigoras estão majoritariamente no Norte do País, enquanto as mais tranquilas estão no Sul e no Sudeste.

Devido à quantidade de variáveis qualitativas ser predominante, podemos avaliar como se dá a relação entre as categorias das variáveis envolvidas no estudo.

O gráfico faz uma redução dimensional com base nas propoções de linhas e colunas através da decomposição espectral (MINGOTI, 2020). Todos esses resultados comprovam o que já havia sido confirmado pelo teste, pelos gráficos e pelas medidas descritivas. Isto é, todas as análises convergem para os mesmos resultados.

Para evitar a poluição visual no gráfico, foi realizada outra análise com unidade federativa e com mês. # Conclusão

A análise e o pré-processamento do dataset datatran_2015_2025 demonstram a importância de compreender a estrutura dos dados antes da inferência ou da análise descritiva simplesmente. O datatran_2015_2025 é longo e desbalanceado, com características predominantemente categóricas com muito confundimento entre as respostas. Para tanto, foi necessário realizar uma etapa de pré-processamento extensa e cansativa para corrigir pelo menos algumas variáveis associadas ao objetivo da pesquisa. Dessa teapa, foi gerado o dataset final data_15_25 apenas com as variáveis consideradas relevantes para o estudo e sem redundâncias semânticas.

A análise exploratória, realizada no data_15_25, revelou relações importantes entre as características. A eficácia de visualizações como gráficos para distinguir os sinistros com vítimas fatais de acordo com as demais variáveis foi fundamental para a pesquisa. Das quatro hipóteses levantadas, três foram validadas, pois uma delas foi se referia à acidentes de motocileta e o banco de dados não continha tal informação:

1. ***A quantidade de óbitos teve tendência positiva de acordo com o ano?*** Não, dado as análises realizadas através de medidas, de gráficos e de modelos, observamos um comportamento parabólico de rau 2 com concavidade positiva (concavidade para cima), indicando uma queda inicial seguida de um aumento;

2. ***Existe uma relação entre acidentes fatais e dia da semana?*** Existe forte relação entre os óbitos e os dias da semana, em especial no final de semana (sexta, sábado e domingo), que foi confirmado pelo gráfico de barras, pelo teste e pela análise de correspondência;

3. ***Há associação entre a condição de visibilidade e a ocorrência de óbito?*** Sim, segundo a análise das medidas, dos gráfico e da análise de correspondência, a visibilidade parcial tem maior incidência na ocorrência de acidentes fatais, sendo a condição boa e ruim relacionadas considerados fatores de proteção.

Além disso, foi detectada a relação com um maior risco de acidente fatal na região norte, seguida pela região nordeste. Tais fatos podem ser devido às condições estruturais das rodovias. Todas as variáveis avaliadas tiveram relação confirmadas estatisticamente ao nível de 5% de significância pelo teste qui-quadrado de Pearson.

Em resumo, a ocorrência de óbito no sinistro tem relação com a fase do dia, em que plena noite e amanhecer são fatores de disco; com o dia da semana, com predominância para domingo, sábado e sexta; com a fase do dia, com a condição de visibilidade, de modo que a parcial é o fator de risco; e com o tipo de pista; em a pista simples é um fator de risco.

Além das relações diretas com as variáveis candidatas a fatores de risco para a ocorrência de óbito, as relações entre as variáveis foram confirmadas. Por exemplo, existe relação entre todas as variáveis entre si combinadas 2 a 2, que observado pelo teste qui-quadrado e pela medida V de Cramér, que apesar de não demonstrarem tanta força, mas têm relação. Tais resultados são relevantes para uma tomada de decisão mais segura para os condutores de modo que possam evitar acidentes fatais, bem como as instituições podem elaborar projetos públicos que conscientizem os condutores.

Como trabalho futuro, pode ser aplicado um modelo de regressão logística e uma análise discriminante usando metodologias de aprendizado supervisionado e usando as metodologias de ciência de dados, para treinamento e validação do modelo, ou até mesmo modelos diferentes para classificação, como redes neurais artificiais.
